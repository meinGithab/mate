```python
import numpy as np
import matplotlib.pyplot as plt
import math

# ==================== INTRARE DATE ====================
print("=" * 60)
print("ANALIZĂ FUNCȚIE DE GRADUL 2: f(x) = ax² + bx + c")
print("=" * 60)

# Parametri - modifica valorile pentru alte funcții
a = 1      # coeficientul lui x^2
b = -5     # coeficientul lui x
c = 6      # termenul liber

print(f"\nFuncția dată: f(x) = {a}x² + {b}x + {c}")

if a == 0:
    print("❌ Nu este o funcție de gradul 2, deoarece a = 0 (ecuație liniară)")
else:
    print("\n" + "=" * 60)
    print("CALCULUL INTERSECȚIILOR CU AXELE")
    print("=" * 60)
    
    # ==================== INTERSECȚIA CU OX ====================
    print("\n📍 INTERSECȚIA CU AXA OX (unde y = 0):")
    D = b**2 - 4 * a * c
    print(f"   Discriminantul: Δ = b² - 4ac = {b}² - 4·{a}·{c} = {D}")
    
    roots = []
    if D > 0:
        x1 = (-b - math.sqrt(D)) / (2 * a)
        x2 = (-b + math.sqrt(D)) / (2 * a)
        roots = [(x1, 0), (x2, 0)]
        print(f"   ✓ Două intersecții diferite:")
        print(f"     - A({x1:.4f}, 0)")
        print(f"     - B({x2:.4f}, 0)")
    elif D == 0:
        x1 = -b / (2 * a)
        roots = [(x1, 0)]
        print(f"   ✓ Intersecție dublă (tangentă la axa Ox):")
        print(f"     - C({x1:.4f}, 0)")
    else:
        print(f"   ✗ Nicio intersecție cu axa Ox (Δ < 0, parabola nu taie Ox)")
    
    # ==================== INTERSECȚIA CU OY ====================
    print("\n📍 INTERSECȚIA CU AXA OY (unde x = 0):")
    y_intercept = c
    print(f"   f(0) = {a}·0² + {b}·0 + {c} = {y_intercept}")
    print(f"   ✓ Intersecție: D(0, {y_intercept})")
    
    # ==================== VÂRFUL PARABOLEI ====================
    print("\n📍 VÂRFUL (EXTREMUL) PARABOLEI:")
    x_v = -b / (2 * a)
    y_v = a * x_v**2 + b * x_v + c
    parabola_type = "minimum" if a > 0 else "maximum"
    print(f"   x_v = -b/(2a) = -{b}/(2·{a}) = {x_v:.4f}")
    print(f"   y_v = f({x_v:.4f}) = {y_v:.4f}")
    print(f"   V({x_v:.4f}, {y_v:.4f}) - punct de {parabola_type}")
    
    # ==================== SENS PARABOLEI ====================
    print("\n📍 ORIENTARE PARABOLĂ:")
    if a > 0:
        print(f"   ✓ a = {a} > 0 → Parabola deschisă în sus (∪)")
    else:
        print(f"   ✓ a = {a} < 0 → Parabola deschisă în jos (∩)")
    
    # ==================== AКСЕ SIMETRIE ====================
    print("\n📍 AXA DE SIMETRIE:")
    print(f"   x = {x_v:.4f}")
    
    # ==================== DETERMINARE INTERVAL ====================
    if roots:
        root_xs = [r[0] for r in roots]
        x_min = min(root_xs + [x_v]) - 2
        x_max = max(root_xs + [x_v]) + 2
    else:
        x_min = x_v - 5
        x_max = x_v + 5
    
    # ==================== DESENARE GRAFIC ====================
    print("\n" + "=" * 60)
    print("REPREZENTAREA GRAFICĂ")
    print("=" * 60)
    
    x = np.linspace(x_min, x_max, 500)
    y = a * x**2 + b * x + c
    
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # Curba funcției
    ax.plot(x, y, 'b-', linewidth=2.5, label=f"f(x) = {a}x² + {b}x + {c}")
    
    # Axele de coordonate
    ax.axhline(0, color='black', linewidth=1.2, alpha=0.7)
    ax.axvline(0, color='black', linewidth=1.2, alpha=0.7)
    ax.grid(True, alpha=0.3, linestyle='--')
    
    # Axa de simetrie
    ax.axvline(x_v, color='green', linewidth=1.5, linestyle='--', alpha=0.6, label=f"Axa simetrie: x = {x_v:.2f}")
    
    # Vârful parabolei
    ax.scatter([x_v], [y_v], color='red', s=200, zorder=5, marker='*', 
              label=f"Vârf V({x_v:.2f}, {y_v:.2f})", edgecolors='darkred', linewidth=2)
    
    # Intersecții cu axa Ox
    if roots:
        x_roots = [r[0] for r in roots]
        ax.scatter(x_roots, [0]*len(x_roots), color='purple', s=150, zorder=5, marker='o',
                  label=f"Intersecții cu Ox: {', '.join([f'({x:.2f}, 0)' for x in x_roots])}", 
                  edgecolors='darkviolet', linewidth=2)
    
    # Intersecție cu axa Oy
    ax.scatter([0], [y_intercept], color='orange', s=150, zorder=5, marker='s',
              label=f"Intersecție cu Oy: (0, {y_intercept})", edgecolors='darkorange', linewidth=2)
    
    # Formatare
    ax.set_xlabel("x", fontsize=12, fontweight='bold')
    ax.set_ylabel("f(x)", fontsize=12, fontweight='bold')
    ax.set_title(f"Graficul funcției f(x) = {a}x² + {b}x + {c}", fontsize=14, fontweight='bold')
    ax.legend(loc='best', fontsize=10, framealpha=0.95)
    ax.set_xlim(x_min, x_max)
    
    plt.tight_layout()
    plt.show()
    
    print("\n✅ Analiza completă a funcției a fost realizată!")
```

```python
# EXEMPLE SUPLIMENTARE DE TESTE
print("\n" + "=" * 60)
print("EXEMPLE SUPLIMENTARE")
print("=" * 60)

# Test 1: Funcție cu 2 rădăcini reale
print("\n🔍 TEST 1: f(x) = 2x² - 8x + 6")
test_a, test_b, test_c = 2, -8, 6
test_D = test_b**2 - 4*test_a*test_c
print(f"   Δ = {test_D} → {'2 rădăcini diferite' if test_D > 0 else '1 rădăcină dublă' if test_D == 0 else 'nicio rădăcină reală'}")
if test_D >= 0:
    if test_D == 0:
        x_test = -test_b / (2*test_a)
        print(f"   x = {x_test}")
    else:
        x_test1 = (-test_b - math.sqrt(test_D)) / (2*test_a)
        x_test2 = (-test_b + math.sqrt(test_D)) / (2*test_a)
        print(f"   x₁ = {x_test1:.4f}, x₂ = {x_test2:.4f}")

# Test 2: Funcție cu rădăcină dublă
print("\n🔍 TEST 2: f(x) = x² - 4x + 4")
test_a, test_b, test_c = 1, -4, 4
test_D = test_b**2 - 4*test_a*test_c
print(f"   Δ = {test_D} → Rădăcină dublă")
x_test = -test_b / (2*test_a)
print(f"   x = {x_test}")

# Test 3: Funcție fără rădăcini reale
print("\n🔍 TEST 3: f(x) = x² + 1")
test_a, test_b, test_c = 1, 0, 1
test_D = test_b**2 - 4*test_a*test_c
print(f"   Δ = {test_D} < 0 → Nicio intersecție cu axa Ox")
print(f"   Parabola nu taie axa Ox (deschisă în sus, minim > 0)")

print("\n💡 Pentru a testa alte funcții, modifică valorile pentru a, b, c în celula anterioară!")
```

# Analiză Completă a Funcției de Gradul 2

## 📋 Instrucțiuni:
- Modifică valorile pentru `a`, `b`, și `c` din codul de mai jos
- Programul va calcula:
  - **Intersecțiile cu axa OX** (rădăcinile ecuației)
  - **Intersecția cu axa OY** (termenul liber)
  - **Vârful parabolei** (coordonatele extremului)
  - **Axa de simetrie** 
  - **Reprezentarea grafică** cu toate punctele speciale marcate

## 📐 Forma funcției:
$$f(x) = ax^2 + bx + c$$

Unde:
- **a** = coeficientul lui x²
- **b** = coeficientul lui x  
- **c** = termenul liber (intersecția cu Oy)
